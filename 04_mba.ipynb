```python
import pandas as pd
# Load the functions from the provided script
from product_focus_mba import load_data, prepare_transaction_data, perform_mba, identify_cross_selling_opportunities, visualize_network

data = load_data(sample=0.2)
```

```python
data.head()
```

```python
print("Loading data...")
data = load_data()

print("Preparing transaction data...")
transactions = prepare_transaction_data(data)

print("Performing Market Basket Analysis...")
rules = perform_mba(transactions, min_support=0.003, min_threshold=0.3)

print("Identifying cross-selling opportunities...")
cross_sell_opportunities = identify_cross_selling_opportunities(rules, top_n=10)

print("\nTop cross-selling opportunities:")
print(cross_sell_opportunities[['antecedents', 'consequents', 'lift', 'confidence']])

print("\nVisualizing network...")
visualize_network(rules, min_lift=1, max_nodes=30)
```
